```python
#to split the fasta file into multiple files and filename = the header of the sequence
infile = open('nd2.fasta', 'r')
outfile = []

for line in infile:
    if line.startswith(">"):
        if (outfile != []): outfile.close()
        genename = line.strip().split('>')[1]
        filename = genename+".fasta"
        outfile = open(filename,'w')
        outfile.write(line)
    else:
        outfile.write(line)
outfile.close()
```

```python
#For observed kmers of the first splited file
file = open('Adelomyia_melanogenys.fasta', 'r')
fasta = file.readlines()
for k in range(1,len(fasta[1])):
    Observed_kmers = {}
    for line_num,line in enumerate(fasta): 
        if line_num == 1: 
            line = line.rstrip() #remove newline \n
            for i,base in enumerate(line[:-k]): 
                kmer = line[i:i+k]  
                if kmer in Observed_kmers:
                    Observed_kmers[kmer]+= 1  
                else:
                    Observed_kmers[kmer] = 1
    print(k,':',len(Observed_kmers))
```
